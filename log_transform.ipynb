```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [16,4]
```

```python
df = pd.read_csv('covid19_globalcases_ts.csv')
df.drop(['Province/State', 'Country/Region', 'Lat', 'Long'], axis = 1, inplace = True)
s = df.sum()
```

```python
def plotcurve(data, title):
    plt.plot(data)
    plt.title(title)
    plt.xlim(data.index.min(), data.index.max())
    plt.ylim(data.min(), data.max())
    plt.plot([data.index.min(), data.index.max()], [data.min(), data.max()], ls="--", c="r")
    plt.xticks(rotation='vertical')
    plt.show()
```

```python
plotcurve(s, 'non-trasnformed values')
```

```python
plotcurve(np.log10(s), 'log base 10')
```

```python
plotcurve(np.log2(s), 'log base 2')
```

```python
plotcurve(np.log(s), 'log base e (natural logarithm)')
```
